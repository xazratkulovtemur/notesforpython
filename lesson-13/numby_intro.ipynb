````python
import numpy as np # type: ignore

arr = np.array([
    [[1, 2, 3], [4, 5, 6]],  # Matrix 1
    [[7, 8, 9], [10, 11, 12]] # Matrix 2
])

print(arr.shape)  # (2, 2, 3) -> (Depth, Rows, Columns)
print(arr)

transposed = arr.transpose(2,1,0)
print(transposed.shape)  # (3, 2, 2)
print(transposed)
````

```
(Depth, Rows, Columns) → (2, 2, 3) Before
(Columns, Rows, Depth) → (3, 2, 2) After
```
