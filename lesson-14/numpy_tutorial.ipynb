````python
import numpy as np
import matplotlib as plt

a=np.array([1,2,3,3,3,4,4,5,19])

def moment(a,k):
    return np.mean((a-a.mean())**k)


skewness = moment(a,3)/(a.std())**3

kurtosis=moment(a,4)/(a.std())**4

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
````

````python
@np.vectorize #helps us to read array without loop
def categorize(x):
    if x < 0:
        return 'negative'
    elif x == 0:
        return 'zero'
    else:
        return 'positive'
categorize(a)
````

````python
#linalg

arr=np.array([[1,2],[3,4]])
arr_inv=np.linalg.inv(arr)

print(arr)
print(arr_inv) 
#inv
#first, 1 and 4 is swapped
#then - is put to 3 and 2

#determinant = 1*4 -2*3 = -2

#-1/2 * [4, - 2]   [-2, 1]
#       [-3,  1]  =[3/2, -1/2]
````

````python
A=np.array([
    [1,2],
    [3,4]
])

B=np.array([
    [1,2],
    [3,4]
])

dot_product1=np.dot(A,B) #first way
print(dot_product1)
dot_product2=A.dot(B) #second way
dot_product3=A@B #third way

print(dot_product2)
print(dot_product3)    
````
