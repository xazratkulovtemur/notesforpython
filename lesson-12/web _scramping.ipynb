````python
import requests


url="https://example.com/news"

response=requests.get(url)

````

````python
from bs4 import BeautifulSoup

soup=BeautifulSoup(response.text, "html.parser") 

title=soup.title.text  #.text is used to make content to be readable
print(f"Page's title: {title}")

paragraph=soup.find('p').text  #find method is used to find specified things

print(f"First paragraph: {paragraph}")
````

````python
#Extracting links

links=soup.find_all('a')

for link in links:
    print(link.get("href"))
````

````python
#Extracting tables

url="https://afd.calpoly.edu/web/sample-tables"
res=requests.get(url)

soup=BeautifulSoup(res.text, "html.parser")

````

````python

tables=soup.find("table")
rows=tables.find_all("tr") #all rows

for row in rows:
    columns=row.find_all("td")
    data=[col.text.strip() for col in columns]  
    print(data)
````

````python
#exracting as dictionary

html = """
<table>
    <tr>
        <th>Name</th>
        <th>Age</th>
        <th>Country</th>
    </tr>
    <tr>
        <td>John</td>
        <td>25</td>
        <td>USA</td>
    </tr>
    <tr>
        <td>Anna</td>
        <td>30</td>
        <td>UK</td>
    </tr>
</table>
"""
soup=BeautifulSoup(html, "html.parser")

table=soup.find("table")

headers=[th.text.strip() for th in table.find_all('th')]

rows=[]

for row in table.find_all('tr')[1:]:
    columns=row.find_all("td")
    row_data={headers[i]: col.text.strip() for i, col in enumerate(columns)}
    rows.append(row_data)

for row in rows:
    print(row)            
````

```

````python

html = """
<div class="content">
    <p id="first">This is the first paragraph.</p>
    <p class="highlight">This is highlighted.</p>
</div> 
"""
soul=BeautifulSoup(html, "html.parser")
soul.find(id="first").text #extracting by id

soul.find(class_="highlight").text  #extracting by class
````

````python
html = '<img src="photo.jpg" alt="Profile Picture">'
soup = BeautifulSoup(html, "html.parser")

image_url = soup.img["src"]
print(image_url)  # Output: photo.jpg
````

````python
html = "<ul><li>Apple</li><li>Banana</li></ul>"
soup = BeautifulSoup(html, "html.parser")

items = [li.text for li in soup.find_all("li")]
print(items)  # Output: ['Apple', 'Banana']
#extracting list items
````

````python
#changing information

from bs4 import BeautifulSoup

html = "<p>Old text</p>"
soup = BeautifulSoup(html, "html.parser")

soup.p.string = "New text"  # Modify the content inside <p>
print(soup)  # Output: <p>New text</p>
````

````python
html= "<p>This is a paragraph.</p>"
soup=BeautifulSoup(html, "html.parser")

soup.p.name="h1"
print(soup) #changing tags
````

````python
html = "<div>Hello</div>"
soup = BeautifulSoup(html, "html.parser")
new_tag=soup.new_tag("h2")
new_tag.string="This is new heading"
soup.div.append(new_tag)
print(soup) 

soup.h2.decompose() #deletes tag
print(soup)
soup.new_tag('h2') #new_tag creates new one
new_tag.string="i am here" #string add string

soup.new_tag('h3')
new_tag.string="Updated info"
soup.h2.replace_with(new_tag)   #replace_with replaces
````

````python
import requests
from bs4 import BeautifulSoup

url='https://data.worldbank.org/country' 
response=requests.get(url)

soup=BeautifulSoup(response.content)
soup
````

````python
countries = {}

sections = soup.find_all('section')
for section in sections:
    title = section.find('h3')
    countries[title.text] = []
    # print(title.text)
    names = section.find_all('a')
    for name in names:
        countries[title.text].append(name.text)

print(countries["U"])
````
