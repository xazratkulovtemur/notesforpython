## Bar charts

```  plt.bar(x, height, width=0.8, color='blue', edgecolor='black')      ```    this is syntaxis of bar chart

``` x: Categories or positions of the bars (e.g., 1, 2, 3).
height: Heights of the bars.
width: Width of the bars (default is 0.8).
color: Fill color of the bars.
edgecolor: Color of the bar edges. 

````python
#basic
catagories = ['Red', 'Blue', 'Black']
values= [4,7,1]

plt.bar(catagories, values, width=0.7, color=[c.lower() for c in catagories], edgecolor='green')
plt.title('Color of clothes')
plt.xlabel("Colors")
plt.ylabel("Quantity")
plt.show()
````

````python
#horizontal
plt.barh(catagories, values, color=[c.lower() for c in catagories], height=0.8)
plt.ylabel("Colors")
plt.xlabel("Values")
plt.title("Colorful clothes")
plt.show()
````

````python
#Grouped bars

catag=[1,2,3,4,5]
values1=[4,5,6,7,8]
values2=[8,7,6,5,4]

x=np.arange(len(catag))

plt.bar(x-0.11, values1, color='b', width=0.2, label="First bar")
plt.bar(x+0.11, values2, color='r', width=0.2, label='Second bar')

plt.title("Grouped bar")
plt.legend() #this will display labels
plt.show()
````

````python
import numpy as np

categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 12]
values2 = [12, 18, 10, 15]

x = np.arange(len(categories))

w = 0.4
plt.bar(x-w/2, values1, width=w, label='Dataset 1', color='blue')
plt.bar(x+w/2, values2, width=w, label='Dataset 2', color='green')

plt.xticks(x, categories) #this will ticks categories to each x value
plt.title('Grouped Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
````

````python
categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 12]
values2 = [5, 10, 3, 8]

plt.bar(categories, values1, label='Dataset 1', color='blue')
plt.bar(categories, values2, label='Dataset 2', color='green')

plt.title('Stacked Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
````

````python
catag1=['A', 'B', 'C']
value1=[1,2,3]
value2=[4,5,6]

plt.bar(catag1, value1, label="data 1", color='Blue')
plt.bar(catag1, value2, label="data 2", color='Green', bottom=value1)

plt.legend()
plt.show()
````

````python
#with different widths

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
colors = ['red', 'blue', 'green', 'purple']
widths = [0.4, 0.6, 0.8, 1.0]

plt.bar(categories, values, width=widths, color=colors, edgecolor="black")
plt.show()
````

# Histograms

``` 
Basic Syntax
plt.hist(data, bins=10, color='blue', edgecolor='black')
data: The input data (a list or NumPy array).
bins: Number of bins or an array defining bin edges.
color: Fill color of the bars.
edgecolor: Color of the bar edges. 

````python
#histogram

import matplotlib.pyplot as plt
import numpy as np

data=np.random.randn(100)
plt.hist(data, bins=20, color='blue', edgecolor="red")
plt.show()
````

````python
#custom histogram

import matplotlib.pyplot as plt
import numpy as np

data=np.random.randn(100)                                       #density=it normalizes data
plt.hist(data, bins=[1,2,3,4,5], color='blue', edgecolor="red", density=True, alpha=0.5) #alpha  = transparency
plt.show()
````

````python
#histogram

import matplotlib.pyplot as plt
import numpy as np

data=np.random.randn(100)
plt.hist(data, bins=20, color='blue', edgecolor="red", cumulative=True) #this adds all values
plt.show()
````

````python
data = np.random.exponential(1, 1000)

plt.hist(data, bins=20, cumulative=True, color='green', edgecolor='black', alpha=0.7)
plt.title('Cumulative Histogram')
plt.xlabel('Value')
plt.ylabel('Cumulative Frequency')
plt.show()
````

````python
data1 = np.random.normal(0, 1, 1000) #0 is mean, 1 is SD, 1000 is range
 
````

````python
data1=np.random.normal(0,1,1000)
data2=np.random.normal(1,1,100)

plt.hist(data1, bins=30, alpha=0.5, label="Mean: 0", color="Blue", density=True, edgecolor="k")
plt.hist(data2, bins=30, alpha=0.5, label="Mean: 1", color="Red", density=True, edgecolor="k")

plt.legend()
plt.show()
````

````python
#uniform

data = np.random.uniform(0, 10, 50000)

plt.hist(data, bins=15, color='cyan', edgecolor='black', alpha=0.8, rwidth=0.9)
plt.title('Customized Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## 3d


````python
#graph
x = np.linspace(-5, 6, 10000)

def f_x(x):
    return 2*x**2 - 3*x + 1

y = f_x(x)

plt.plot(x, y, label='$f(x) = 2x^2 - 3x + 1$')
plt.title('Parabola')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(10, 3))

x = np.linspace(-5, 5, 10000)
y = 3 * np.sin(3*x + np.pi)
plt.plot(x, y, label=r"$f(x) = 3\sin(3x + \pi)$")
plt.legend(loc='upper left')
plt.show()
````

````python
x = np.linspace(-5, 5, 10000)
y = np.where(x<0, x**2, np.sin(x))


plt.plot(x, y)
plt.show()
````
